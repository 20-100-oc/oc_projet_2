```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

from sklearn.neighbors import LocalOutlierFactor
```

```python
data_path = 'G:/OpenClassrooms/projet 2/p2-arbres-fr.csv'
data = pd.read_csv(data_path, sep=';')
```

```python
print(data.shape)
dtypes = set([data[column].dtype for column in data.columns])
print(dtypes)
```

```python
def delete_multivariate_outliers(dataframe):
    # delete outlier 5%
    lof = LocalOutlierFactor(n_neighbors=5, n_jobs=-1)
    lof.fit_predict(dataframe.select_dtypes(['float64', 'int32']).dropna())
    indices = dataframe.select_dtypes(['float64', 'int32']).dropna().index
    df_lof = pd.DataFrame(index=indices,
                          data=lof.negative_outlier_factor_, columns=['lof'])
    index_to_drop = df_lof[df_lof['lof'] < np.quantile(
        lof.negative_outlier_factor_, 0.05)].index
    return dataframe.drop(index_to_drop, axis=0)
```

```python
columns_to_keep = ['hauteur_m', 'circonference_cm', 'stade_developpement', 'genre']

temp_data = data[columns_to_keep]
print(data.shape)
print(temp_data.shape)
temp_data.head()
```

```python
clean_data = delete_multivariate_outliers(temp_data)
print(clean_data.shape)
clean_data.head()
```
